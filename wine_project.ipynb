```python
df0=white_wine.drop(['chlorides', 'volatile acidity', 'fixed acidity', 'citric acid', 'free sulfur dioxide',
'density', 'pH', 'quality','quality_lable'], axis=1)
df0['volatile acidity']=white_wine['volatile acidity']
df0['quality']=white_wine['quality']

df1=red_wine.drop(['chlorides', 'volatile acidity', 'fixed acidity', 'citric acid', 'free sulfur dioxide',
'density', 'pH', 'quality','quality_lable'], axis=1)
df1['volatile acidity']=red_wine['volatile acidity']
df1['quality']=red_wine['quality']
pd.concat([df1.describe(), df0.describe()], axis=1, keys=['red wine status','white wine status'])
# 1) residual suger is higher in white wine, std is also higher the same is total sulfur dioxide
# 2) quality of white wines is higher
# 3)  mean value of sulfates and volatile acidity seem to be higher in red wine as compared to white wine.
# 4) fixed acidity, choloride  is higher in red wine
# 5) citric acid, 	ree sulfur dioxide	total sulfur dioxide of white wine are higher.
# 6) pH	sulphates of red wine is higher
# 7) median of residual suger for white wine is almost double of red wine. while its maximum is 
# nearly 4 time more
# 8) In the U.S, the legal limits of Volatile Acidity are 1.2 g/L for red table wine 
# and 1.1 g/L for white table wine.
# It is nice to see how many percent of data are folling this US stabdard
# 9) Chlorides can be a significant contributor to saltiness in wine. (find the satiest wines)
# 10) the part of the sulfur dioxide that is added to a wine and that is lost into it is said to be 
# bound, while the active part is said 
# to be free. The winemaker will always try to get the highest proportion of free sulfur to bind. 
# to find how much  sulfur dioxide  is bound.
# 11)Total sulfur dioxide is the sum of the bound and the free sulfur dioxide (SO2). Here, 
# its expressed in mg/dm3. There are legal limits for sulfur levels in wines: in the EU, red wines 
# can only have 160mg/L, while white and rose wines can have about 210mg/L. Sweet wines are allowed 
# o have 400mg/L. For the US, the legal limits are set at 350mg/L, and for Australia, this is 250mg/L.
# compare the data with standards
# Density is generally used as a measure of the conversion of sugar to alcohol
# see the cov (density and alcohol)
# 12) pH or the potential of hydrogen is a numeric scale to specify the acidity or
#  basicity the wine. As you might know, solutions with a pH less than 7 are acidic, 
# while solutions with a pH greater than 7 are basic. With a pH of 7, pure water is neutral. 
# Most wines have a pH between 2.9 and 3.9 and are therefore acidic.
# compare the ph with the values give above
# 14) Sulfates are to wine as gluten is to food. You might already know sulfites from the 
# headaches that they can cause. They are a regular part of winemaking around the world 
# and are considered necessary. In this case, they are expressed in g(potassiumsulphate)/dm3.
# 15) Alcohol: wine is an alcoholic beverage, and as you know, the percentage of 
# alcohol can vary from wine to wine. It shouldnt be surprised that this variable 
# is included in the data sets, where its expressed in % vol.
```
