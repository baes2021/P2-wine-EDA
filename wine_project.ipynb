```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_lable']=red_wine['quality'].apply(lambda x : 'low' if x <= 5 else 'medium' if x <= 7 else 'high')
# here we are transforming these labels into categrical data type (specific to pandas) instead of simple st

red_wine['quality_lable']=pd.Categorical(red_wine['quality_lable'], categories=['low', 'medium', 'high'])
red_wine
plt.bar( red_wine['quality_lable'].value_counts().index, red_wine['quality_lable'].value_counts())
sns.barplot( red_wine['quality_lable'].value_counts().index, red_wine['quality_lable'].value_counts())
```

```python
white_wine['quality_lable']=white_wine['quality'].apply(lambda x: 'low' if x<=5 else 'medium' if x <= 7 else 'high')
white_wine['quality_lable']=pd.Categorical(white_wine['quality_lable'], categories= ['low', 'medium','high'])
```

```python
sns.barplot(white_wine['quality_label'].value_counts().index, white_wine['quality_label'].value_counts())
```
