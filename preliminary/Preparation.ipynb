```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
white_wine=pd.read_csv('../practice/winequality-white.csv' , sep=';')
red_wine=pd.read_csv('../practice/winequality-red.csv' , sep=';')
```

```python
white_wine.columns= white_wine.columns.str.replace(' ','_')
red_wine.columns= white_wine.columns.str.replace(' ','_')
```

```python
# outliers
# first plot all data to see how the data looks like
nrow = 3
ncolum=4
count=0
fig=plt.figure(figsize=(12,8))
for i in range(nrow):
    for j in range(ncolum):
        plt.subplot2grid((nrow,ncolum), (i,j))
        sns.histplot(white_wine[white_wine.columns[count]], bins=20, color='white',  edgecolor='black', linewidth=1.0)
        plt.ylabel('frequency')
        count +=1
        plt.tight_layout(pad=2)
```

```python
for i in range(12):
    print('The skewness of {} in white wine is: '.format(white_wine.columns[i]), white_wine[white_wine.columns[i]].astype(float).skew())
```

```python
# log transformation for correcting skweness
df_w=white_wine.copy()
for j in range(2):
    df_w[df_w.columns[j]] = white_wine[white_wine.columns[j]].map(lambda i: np.log(i) if i > 0 else 0) 
for j in range(3,6):
    df_w[df_w.columns[j]] = white_wine[white_wine.columns[j]].map(lambda i: np.log(i) if i > 0 else 0) 
df_w[df_w.columns[7]] = white_wine[white_wine.columns[7]].map(lambda i: np.log(i) if i > 0 else 0)
for i in range(12):
    print('The skewness of {} in white wine is: '.format(df_w.columns[i]), df_w[df_w.columns[i]].astype(float).skew())
```

```python
# outliers
# first plot all data to see how the data looks like
nrow = 3
ncolum=4
count=0
fig=plt.figure(figsize=(12,8))
for i in range(nrow):
    for j in range(ncolum):
        plt.subplot2grid((nrow,ncolum), (i,j))
        sns.histplot(df_w[df_w.columns[count]], bins=20, color='white',  edgecolor='black', linewidth=1.0)
        plt.ylabel('frequency')
        count +=1
        plt.tight_layout(pad=2)
```

```python
# Quantile-based Flooring and Capping

df_rp=red_wine.copy()
for i in range(12):
    df_rp[df_rp.columns[i]] = np.where(red_wine[red_wine.columns[i]] < red_wine[red_wine.columns[i]].quantile(0.01) , 
    red_wine[red_wine.columns[i]].quantile(0.01) , red_wine[red_wine.columns[i]])
    df_rp[df_rp.columns[i]] = np.where(red_wine[red_wine.columns[i]] > red_wine[red_wine.columns[i]].quantile(0.99) , 
    red_wine[red_wine.columns[i]].quantile(0.99) , red_wine[red_wine.columns[i]])
```

```python
for i in range(12):
    print('The skewness of {} in red wine is: '.format(df_rp.columns[i]), df_rp[df_rp.columns[i]].astype(float).skew())
```
