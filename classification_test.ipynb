```python
import numpy as np
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot
X, y = make_blobs(n_samples=2000, centers=3, random_state=1)
print(X.shape, y.shape)
counter = Counter(y)
print(counter)
for i in range(10):
  print(X[i], y[i])
for label, _ in counter.items():
  row_ix = np.where(y == label)[0]
  pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

```python
# above example is a multi-class with one lable
# below is an example of multi-class (4 class) with multi-lable (4 label). For this we have to use Multi-label Random Forestsm Multi-label Decision trees,
#Multi-label Gradient Boosting, ..
from sklearn.datasets import make_multilabel_classification
X, y = make_multilabel_classification(n_samples=1000, n_features=3, n_classes=4, n_labels=4, random_state=1)
print(X.shape, y.shape)
for i in range(10):
	print(X[i], y[i])
```
